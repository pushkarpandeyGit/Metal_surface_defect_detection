```python
#importung modules
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
```

```python
#storing the path of train and test dataset
train_dir = "neu_surface_defect/NEU-DET/train/images"
val_dir   = "neu_surface_defect/NEU-DET/validation/images"
```

```python
#uploading the dataset
IMG_SIZE = (128, 128)
BATCH_SIZE = 32

train_ds = tf.keras.utils.image_dataset_from_directory(train_dir,image_size=IMG_SIZE,batch_size=BATCH_SIZE)

val_ds = tf.keras.utils.image_dataset_from_directory(val_dir,image_size=IMG_SIZE,batch_size=BATCH_SIZE,shuffle=False )

class_names = train_ds.class_names

#checking the classess in dataset
print("Classes:", class_names)
```

```python
#optimizing dataset for reducing overfitting
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.prefetch(AUTOTUNE)
val_ds   = val_ds.prefetch(AUTOTUNE)

```

```python
#data augmentation for reducing overfitting
data_augmentation = tf.keras.Sequential([layers.RandomFlip("horizontal"),layers.RandomRotation(0.2),layers.RandomZoom(0.2),layers.RandomContrast(0.2),layers.RandomTranslation(0.1, 0.1)])
```

```python
#building the trained model
base_model = MobileNetV2(input_shape=(128,128,3),include_top=False,weights='imagenet')

base_model.trainable = False
```

```python
#building our model ( giving input processed in from of input to MobileNetV2)
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
model = models.Sequential([layers.Input(shape=(128,128,3)),data_augmentation,layers.Lambda(preprocess_input), base_model,layers.GlobalAveragePooling2D(),
    layers.Dropout(0.5),
    layers.Dense(6, activation='softmax')])
```

```python
#compiling the mdoel
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

```

```python
model.summary()
```

```python
#earlt stopping for reducing overfitting
early_stop = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)

#training data to model
history = model.fit(train_ds,validation_data=val_ds,epochs=50,callbacks=[early_stop])
```

```python
plt.figure(figsize=(12,5))

# plotting accuracy
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title('Accuracy')
plt.legend()

# plotting loss
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title('Loss')
plt.legend()

plt.show()
```

```python
base_model = model.layers[2]
gap = model.layers[3]
drop = model.layers[4]
dense = model.layers[5]
```

```python
import tensorflow as tf

deploy_model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(128,128,3)),
    tf.keras.layers.Rescaling(1./127.5, offset=-1),
    base_model,
    gap,
    drop,
    dense
])
```

```python
deploy_model.summary()

```

```python
deploy_model.save("neu_defect_deploy.keras")
```

## Confusion matrix

```python
import numpy as np

# True labels
y_true = np.concatenate([y for x, y in val_ds], axis=0)

# Model predictions
y_pred_probs = model.predict(val_ds)
y_pred = np.argmax(y_pred_probs, axis=1)

class_names = [
    'crazing',
    'inclusion',
    'patches',
    'pitted_surface',
    'rolled-in_scale',
    'scratches'
]

```

```python
#creating cinfusion matrix
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
cm = confusion_matrix(y_true, y_pred)

#plotting heatmap for actual vs predicted output
plt.figure(figsize=(8,6))
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues',xticklabels=class_names,yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix  NEU Surface Defect")
plt.show()
```

```python
print(classification_report(y_true, y_pred, target_names=class_names))
```
